```r
# This code is originally from Guillaume Wilemme's Lecture on Intro to Structural Econometrics

# clear memory
rm(list = ls())
set.seed(123456789)

# packages
# install.packages("optimx")
library(optimx)
```

```r
# The code builds on the model by Rust (1987)
# Denote that i = 1 for no replacement, and i = 2 for replacement
# x is the mileage per month
# (x,i) pair is observed by harold zurcher (as in Rust (1987))

## Constant in the model
beta = 0.95   # beta is not estimated
xgrid <- seq(0,5,0.3)   # approximate space of x by a grid
```

```r
# Auxiliary Functions

evalEV <- function(x, i, EV) {
    ## EV is a matrix
    ## to evaluate EV at (x,i)
    ### Use function match(). Example below (so that I don't forget!)
#     > v1 <- c("a1","b2","c1","d2")
#     > v2 <- c("g1","x2","d2","e2","f1","a1","c2","b2","a2")
#     > x <- match(v1,v2)
#     > x
#     [1] 6 8 NA 3
    
    return(EV[match(x,xgrid),i])
}

discretize <- function(cdffun) {
    ## This function creates a discretized probability function on xgrid given a continuous CDF
    ## Remember that xgrid = [x1 x2 ... xN]
    
    ## First, initialize output pgrid. pgrid is a vector with the same size as xgrid
    pgrid <- rep(0, length(xgrid))
    
    ## p(x1) = F(x1) with F being the cdffun
    pgrid[1] <- cdffun(xgrid[1])
    
    ## Now iterate
    for (i in 2:(length(xgrid) - 1)) {
        
        ## p(xi) =  F(x(i+1)) - F(xi)
        pgrid[i] <- cdffun(xgrid[i+1]) - cdffun(xgrid[i])
    }
    
    ## p(xN) = 1 - F(xN)
    pgrid[length(xgrid)] <- 1 - cdffun(xgrid[length(xgrid)])
    
    ## return
    return(pgrid)
    
}

draw_discretize <- function(cdffun) {
    ## Generate a draw from the discretized distribution
    
    pgrid <- discretize(cdffun)
    
    ## return
    ### REMEMBER: rmultinom(n, size, prob)
    ### Generate multinomially distributed random number vectors and compute multinomial probabilities.
    ### example: rmultinom(n = 10, size = 12, prob = c(0.1,0.2,0.8))
    
    return(sum( rmultinom(1,1,pgrid) * xgrid ))

}

```

```r
# Deterministic Model - Simulation

utility <- function(x, i, theta) {
    
    # page 20 of the slide session 2
    
    if (i == 1) {
        return(-x * theta$t1)
    }
    else {
        # if (i == 2)
        return(-theta$RC - 0*theta$t1)
    }
    
}

cdf1 <- function(x1, x0, i, theta) {
    
    # p(x1\x0,i,theta), we take the exponential
    ## Density, distribution function, quantile function and 
    ## random generation for a generalisation of the exponential distribution, 
    ## in which the rate changes at a series of times.
    ## pexp(q, rate = 1, lower.tail = TRUE, log.p = FALSE)
    
    if (i == 1) {
        return(pexp(x1 - x0, theta$t2))
    }
    else {
        return(pexp(x1,theta$t2))
    }
}

model1_EV <- function(theta) {
    ## DGP of the deterministic model
    
    maxutility <- function(x, EV, theta) {
        ## Check Equation 15 of SLide Session 2 page 21/26
        return( max( utility(x,1,theta) + beta*evalEV(x,1,EV), utility(x,2,theta) + beta*evalEV(x,2,EV) ) )
    }
    
    transformation <- function(EV,theta) {
        ## Initialization
        EVoutput <- EV
        
        for (xindex in seq_along(xgrid)) {
            
            ## Why seq_along() instead of seq()?
            ## Check: https://stackoverflow.com/questions/13732062/what-are-examples-of-when-seq-along-works-but-seq-produces-unintended-results
            
            maxvector <- sapply(xgrid,function(x1){ maxutility(x1, EV, theta) })
            
            for (i in c(1,2)) {
                pgrid <- discretize(function(x1){cdf1(x1, xgrid[xindex], i, theta)}) 
                EVoutput[xindex,i] <- maxvector %*% pgrid   ## scalar product
            }
        }
        
        ## return
        return(EVoutput)
        
    }
    
    EV <- matrix(0,nrow=length(xgrid),ncol=2)
    
    max_err <- 100
    nloop <- 0
    
    ## NOTE!!
    ## The strategy here is to PRAY (as we did not show it) that the transformation is a Contraction Mapping
    ## Begin Contraction Mapping
    
    while (max_err > 1e-14) {
        nloop  <- nloop + 1
        EV2 <- transformation(EV,theta)
        max_err <- max(EV - EV2)
        EV <- EV2
#         print(max_err)
    }
#     print(nloop)
    return(EV)
}
```

```r
# Stochastic Model - Simulation

model2_EV = function(theta){
  ## dgp of deterministic model
  lse_utility = function(x,EV,theta){
    exp1 = exp( utility(x,1,theta)+beta*evalEV(x,1,EV) )
    exp2 = exp( utility(x,2,theta)+beta*evalEV(x,2,EV) )
    return(log(exp1+exp2))
  }
  transformation = function(EV,theta){
    EVoutput = EV ##initialization
    for (xindex in seq_along(xgrid)) {
      lse_vector = sapply(xgrid,function(x1){lse_utility(x1,EV,theta)})
      for (i in c(1,2)){
        pgrid = discretize(function(x1){cdf1(x1,xgrid[xindex],i,theta)}) 
        EVoutput[xindex,i] = lse_vector %*% pgrid ## scalar product
      }
    }
    return(EVoutput)
  }
  EV = matrix(0,nrow=length(xgrid),ncol=2)
  max_err = 100
  # nloop=0
  while (max_err>0.01) {
    # nloop =nloop+1
    EV2 = transformation(EV,theta)
    max_err = max(EV - EV2 )
    EV = EV2
    #     print(max_err)
  }
  #  print(nloop)
  return(EV)
}
model2_Popt2 = function(x,EV,theta){
  ## compute the probability to draw i=2
  exp1 = exp( utility(x,1,theta)+beta*evalEV(x,1,EV) )
  exp2 = exp( utility(x,2,theta)+beta*evalEV(x,2,EV) )
  return(exp2/(exp1+exp2))
}
model2_draw1 = function(TT,EV,theta){
  ### simulate 1 observations over TT periods
  x = rep(0,TT)
  i = rep(0,TT)
  ### initialize x1 and i1
  x[1] = draw_discretize(function(xx){cdf1(xx,0,1,theta)}) 
  i[1] = 1 + rbinom(1,1,model2_Popt2(x[1],EV,theta))
  ### loop
  for (tt in 2:TT){
    x[tt] = draw_discretize(function(xx){cdf1(xx,x[tt-1],i[tt-1],theta)}) 
    i[tt] = 1 + rbinom(1,1,model2_Popt2(x[tt],EV,theta))
  }
  return(list(x=x,i=i))
}
model2 = function(N,TT,theta){
  ### simulate N observations over TT periods
  xmatrix = matrix(0,nrow=N,ncol=TT) ## xmatrix[m,tt] is the value of x for the bus m and time tt
  imatrix = matrix(0,nrow=N,ncol=TT)
  EV = model2_EV(theta)
  for (m in 1:N){
    obs = model2_draw1(TT,EV,theta)
    xmatrix[m,] = obs$x
    imatrix[m,] = obs$i
  }
  return(list(xmat=xmatrix,imat=imatrix))
}
```

```r
# Stochastic Model - Estimation

loglike_p1_m = function(x,i,theta){
  ## individual (log) partial likelihood, x and i are two vectors of length TT
  ## it is the proba to obtain the sequence x, taking i as given
  TT = length(x)
  ### deal with x1
  index = match(x[1],xgrid) 
  pgrid = discretize(function(xx){cdf1(xx,0,1,theta)}) 
  proba = pgrid[index]
  ### loop
  for (tt in 2:TT){
    index = match(x[tt],xgrid) 
    pgrid = discretize(function(xx){cdf1(xx,x[tt-1],i[tt-1],theta)}) 
    proba = proba * pgrid[index]
  }
  return(log(proba))
}
loglike_p1 = function(data,theta){
  ## individual (log) partial likelihood, x and i are two vectors of length TT
  N = length(data$x[,1])
  output = 0
  for (m in 1:N){
    output = output + loglike_p1_m(data$x[m,],data$i[m,],theta)
  }
  return(output) 
}
loglike_p2_m = function(x,i,EV,theta){
  ## individual (log) partial likelihood, x and i are two vectors of length TT
  ## it is the proba to obtain the sequence i, taking x as given
  TT = length(x)
  proba = 1
  ### loop
  for (tt in 1:TT){
    if (i[tt]==2){
      proba = proba * model2_Popt2(x[tt],EV,theta)
    }
    else {
      proba = proba * (1-model2_Popt2(x[tt],EV,theta))
    }
  }
  return(log(proba))
}
loglike_p2 = function(data,theta){
  ## individual (log) partial likelihood, x and i are two vectors of length TT
  N = length(data$x[,1])
  EV = model2_EV(theta)
  output = 0
  for (m in 1:N){
    output = output + loglike_p2_m(data$x[m,],data$i[m,],EV,theta)
  }
  return(output)
}
loglike_full = function(data,theta,xgrid){loglike_p1(data,theta)+loglike_p2(data,theta)}

estim1 = function(data,guessedtheta){
  ## first stage of estimation, estimate for theta2
  obj = function(theta2){## log partial likelihood as a function of theta2
    stheta = list(RC=guessedtheta$RC,t1=guessedtheta$t1,t2=theta2) ##the values of RC and t1 do not matter
    return(-loglike_p1(data,stheta))
  }
  theta2_est = optimx(guessedtheta$t2,obj,control=list(method=c("Nelder-Mead","BFGS","nlm"),trace=1))
  return(theta2_est)
}
estim2 = function(data,guessedtheta){
  ## second stage of estimation, estimate for RC and theta1
  obj = function(v){## log partial likelihood as a function of the vector [RC,theta1]
    stheta = list(RC=v[1],t1=v[2],t2=guessedtheta$t2) ##the values of RC and t1 do not matter
    return(-loglike_p2(data,stheta))
  }
  RCtheta1_est = optimx(c(guessedtheta$RC,guessedtheta$t1),obj,method="BFGS",lower=c(0,0),upper=c(20,3),control=list(trace=1))
  return(RCtheta1_est)
}
estimfull = function(data,guessedtheta){
  ## full estimation, estimate for RC, theta1 and theta2
  obj = function(v){## log partial likelihood as a function of the vector [RC,theta1]
    stheta = list(RC=v[1],t1=v[2],t2=v[3]) ##the values of RC and t1 do not matter
    return(-loglike_p1(data,stheta)-loglike_p2(data,stheta))
  }
  #  theta_est = optimx(c(guessedtheta$RC,guessedtheta$t1,guessedtheta$t2),obj,method="BFGS",lower=c(0,0,0),upper=c(20,3,5),control=list(trace=1))
  theta_est = optimx(c(guessedtheta$RC,guessedtheta$t1,guessedtheta$t2),obj,method="Nelder-Mead",control=list(trace=1))
  return(theta_est)
}
```

```r
## MAIN: Deterministic Model

theta <- list(RC=10, t1=1, t2=3)
EV <- model1_EV(theta)

# 1. Look at the functions

# 2. Search for the threshold mileage. Plot the expected value of choosing i = 1 and i = 2
## Lifetime value if choice is i == 1 (no replacement)
Vopt1 <- sapply(xgrid, function(x){ utility(x,1,theta) + beta*evalEV(x,1,EV) })

## Lifetime value if choice is i == 2 (replacement)
Vopt2 <- sapply(xgrid, function(x){ utility(x,2,theta) + beta*evalEV(x,2,EV) })

## Plot
plot(xgrid, Vopt1)
lines(xgrid, Vopt2)

```

```r
# MAIN: Stochastic Model

# 3. Code the equivalent of model1_EV but in the stochastic model

# 4. Code a function to simulate N observations of TT periods
data <- model2(300,30,theta)

# 5. We now want to estimate the model by Maximum Likelihood
# Write the partial likelihood function
loglike_p1(data,theta)

loglike_p2(data,theta)

loglike_full(data,list(RC=10,t1=1,t2=3))

loglike_full(data,list(RC=10,t1=1,t2=2))
```

```r
# 6. Try to estimate the model
# This is a good way to experiment an ungrateful part of the job: numerical optimization

guessedtheta <- list(RC=10,t1=1,t2=3)

theta2_est <- estim1(data,guessedtheta)
estim2(data,guessedtheta)

estimfull(data,guessedtheta)
```
